```python
from firebase_admin import firestore
db = firestore.client()
```

```python
db.collection(u'TheMacroDiet/Users/Configuration')
```

```python
userIds = [] 
users = db.collection(u'TheMacroDiet/Production/Users').stream()

for user in users:
    userIds.append(user.id)
    #print(f'{user.id} => {user.to_dict()}')

print(userIds)
```

```python
for userId in userIds:
    print(db.collection(u'TheMacroDiet/Production/Users').document(userId).get().to_dict())
```

```python
for userId in userIds:
    print("For user: " + userId)
    print("--------------------------------------\n")
    userDailyEntries = db.collection(u'TheMacroDiet/Production/Users/'+userId+'/DailyEntries').stream()
    for userDailyEntry in userDailyEntries:       
        entries = db.collection(u'TheMacroDiet/Production/Users/'+userId+'/DailyEntries/'+userDailyEntry.id+'/Entries').stream()
        entriesSize = 0
        for entry in entries:
            entriesSize += 1
        print(userId +", "+ userDailyEntry.id +", "+str(entriesSize))
        #print(userDailyEntry.to_dict())
        entry_ref = db.collection(u'TheMacroDiet/Production/Users/'+ userId +'/DailyEntries').document(userDailyEntry.id)
        entry_ref.set({
            u'SizeOfEntries': entriesSize
        }, merge=True)
```
