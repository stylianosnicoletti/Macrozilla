```python
import firebase_admin
from firebase_admin import credentials

cred = credentials.Certificate('themacrodiet-s-firebase-adminsdk-f0h78-30db3d9e86.json')
default_app = firebase_admin.initialize_app(cred)

```

```python
from firebase_admin import firestore
db = firestore.client()
```

```python
db.collection(u'TheMacroDiet/Users/Configuration')
```

```python
userIds = [] 
users = db.collection(u'TheMacroDiet/Production/Users').stream()

for user in users:
    userIds.append(user.id)
    #print(f'{user.id} => {user.to_dict()}')

print(userIds)
```

```python
for userId in userIds:
    print(db.collection(u'TheMacroDiet/Production/Users').document(userId).get().to_dict())
    print(db.collection(u'TheMacroDiet/Production/Users').document(userId).get().to_dict()['Sizes']['DailyEntries'])
```

```python
for userId in userIds:
    print("For user: " + userId)
    print("--------------------------------------\n")
    user_ref = db.collection(u'TheMacroDiet/Production/Users').document(userId)
    defaultUserData = {
        u'Options': {
             u'DarkMode': False,
             u'UseOnlyPersonalDb': False,
             u'TransferEntriesEnabled': True,
             u'DeletingAllDailyEntriesEnabled': True,

        },
        u'Sizes': {
            u'DailyEntries': 0,
        }
    }
    defaultUserData['Sizes']['DailyEntries'] = user_ref.get().to_dict()['Sizes']['DailyEntries']
    user_ref.set(defaultUserData, merge=True)
    print(defaultUserData)
    print("\n")
```
